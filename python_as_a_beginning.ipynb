```python
# Correlation coefficients and p values for all pairs of rows of a matrix
```

```python
import numpy as np
from scipy.stats import pearsonr

AK = [761,9.0,9.1,14.3]
AL = [780,11.6,17.4,11.5]
AR = [593,10.2,20.0,10.7]
AZ = [715,8.6,15.4,12.1]
matrix = np.array([AK,AL,AR,AZ])#Use np.array rather than np.matrix because pearsonr expects two 1D arrays of length ≥ 2.

def corrcoef_loop(X):

    rows,cols = X.shape[0], X.shape[1]
    
    r = np.ones(shape = (rows, rows))
    p = np.ones(shape = (rows, rows))#Compute row correlations, you can change to cols whenever you want.
    
    for i in range(rows):
        for j in range(i+1, rows):#i+1 because we don't need to compute the correlations between the same row(which is 1 obviously). 
            r_, p_ = pearsonr(X[i, :], X[j, :])#pearsonr(x,y) returns Pearson correlation coefficient and two-tailed p-value for testing non-correlation betweeen rows x and y.
            r[i, j] = r[j, i] = r_
            p[i, j] = p[j, i] = p_#Define each entry in r,p, the above line only returns the upper-triangle of the matrix.
    
    return r, p

print("The correlation matrix and P-values are:", corrcoef_loop(matrix))
```

```python
# Insertion Sort
```

```python
def insertionSort_python(x):
    for i in range(1, len(x)):
        token = x[i]
        j = i - 1
        while j >= 0 and x[j] > token:
            x[j + 1] = x[j]
            j -= 1
        x[j + 1] = token
    return x
```

```python
my_list = [2,5,3,1,4,6]
sorted_list = insertionSort_python(my_list)
print(sorted_list)
```

```python
# Bar chart with individual bar colors
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

micro_planets = ['Ceres','Juno','Pallas','Vesta']
counts = [3,5,8,14]
bar_labels = ['orange','purple','_purple','red']#"_" will hide the second "purple" in the legend. The only legal colours in Matplotlib's tab are blue,orange,green,red,purple,brown,pink,gray,olive and cyan.
bar_colors = ['tab:orange','tab:purple','tab:purple','tab:red']

for planet, count, label, color in zip(micro_planets, counts, bar_labels, bar_colors):#This line can make the legend show separate colours, or they will be treated as a single string.
    ax.bar(planet, count, label = label, color = color)

ax.set_ylabel('Population')
ax.set_title('Population on each micro planet')
ax.legend(title = 'Micro planets')

plt.show()
```

```python
# Stacked bar chart with labels
import matplotlib.pyplot as plt
import numpy as np

distance_from_the_base = ('Same location', '1 to 100', '101 to 200', '201 to 300')
age_expected_frequencies = {
    '0 to 14': np.array([5.196, 3.541, 3.815, 4.251]),
    '15 to 28': np.array([0.635, 0.735, 1.829, 1.71]),
    '29 to 42': np.array([0.846, 0.751, 0.601, 0.629]),
    '43 to 56': np.array([0.616, 0.913, 0.725, 0.759])
}
width = 0.6 # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()
bottom = np.zeros(4)

for age, age_expected_frequency in age_expected_frequencies.items():
    p = ax.bar(distance_from_the_base, age_expected_frequency, width, label = age, bottom = bottom)
    bottom += age_expected_frequency
    
    ax.bar_label(p, label_type='center')
    
ax.set_title('Observed over mean expected frequencies of street robbery in Darlington by distance from the base')
ax.legend()

plt.show()
```

```python
# Grouped bar chart with labels
import matplotlib.pyplot as plt
import numpy as np

agency_name = ("Rainy, Stormy, Cloudy", "Bella Italia", "Fly Low", "Kangeroos", "Asia By Plane")
number = {
    'number of bookings': (27870, 27866, 27869, 27867, 27416),
    'daily count': (4108, 4038, 4037, 4095, 4004),
}

x = np.arange(len(agency_name))# the label locations
width = 0.25 # the width of the bars
multiplier = 0 # first book goes in position 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in number.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label = attribute)
    ax.bar_label(rects, padding = 3)# Place the text label 3 points (3/72 inches) above the top of each bar.
    multiplier += 1 # next book goes one slot to the right, etc.
    
ax.set_ylabel('Number')
ax.set_title('A SELECT statement on the column view')
ax.set_xticks(x + width, agency_name, rotation = 90) # For every item in agency_name, shift one width to the right. Make the labels as vertical to avoid overlapping.
ax.legend(loc='upper left', ncols=2)
ax.set_ylim(0, 40000)

plt.show()
```
