```python
# Correlation coefficients and p values for all pairs of rows of a matrix
```

```python
import numpy as np
from scipy.stats import pearsonr

AK = [761,9.0,9.1,14.3]
AL = [780,11.6,17.4,11.5]
AR = [593,10.2,20.0,10.7]
AZ = [715,8.6,15.4,12.1]
matrix = np.array([AK,AL,AR,AZ])#Use np.array rather than np.matrix because pearsonr expects two 1D arrays of length ≥ 2.

def corrcoef_loop(X):

    rows,cols = X.shape[0], X.shape[1]
    
    r = np.ones(shape = (rows, rows))
    p = np.ones(shape = (rows, rows))#Compute row correlations, you can change to cols whenever you want.
    
    for i in range(rows):
        for j in range(i+1, rows):#i+1 because we don't need to compute the correlations between the same row(which is 1 obviously). 
            r_, p_ = pearsonr(X[i, :], X[j, :])#pearsonr(x,y) returns Pearson correlation coefficient and two-tailed p-value for testing non-correlation betweeen rows x and y.
            r[i, j] = r[j, i] = r_
            p[i, j] = p[j, i] = p_#Define each entry in r,p, the above line only returns the upper-triangle of the matrix.
    
    return r, p

print("The correlation matrix and P-values are:", corrcoef_loop(matrix))
```

```python
# Insertion Sort
```

```python
def insertionSort_python(x):
    for i in range(1, len(x)):
        token = x[i]
        j = i - 1
        while j >= 0 and x[j] > token:
            x[j + 1] = x[j]
            j -= 1
        x[j + 1] = token
    return x
```

```python
my_list = [2,5,3,1,4,6]
sorted_list = insertionSort_python(my_list)
print(sorted_list)
```

```python
# Grouped bar chart with labels
## Scenario: SAP HANA Calculation View

import matplotlib.pyplot as plt
import numpy as np

agency_name = ("Rainy, Stormy, Cloudy", "Bella Italia", "Fly Low", "Kangeroos", "Asia By Plane")
number = {
    'number of bookings': (27870, 27866, 27869, 27867, 27416),
    'daily count': (4108, 4038, 4037, 4095, 4004),
}

x = np.arange(len(agency_name))# the label locations
width = 0.25 # the width of the bars
multiplier = 0 # first book goes in position 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in number.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label = attribute)
    ax.bar_label(rects, padding = 3)# Place the text label 3 points (3/72 inches) above the top of each bar.
    multiplier += 1 # next book goes one slot to the right, etc.
    
ax.set_ylabel('Number')
ax.set_title('A SELECT statement on the column view')
ax.set_xticks(x + width, agency_name, rotation = 90) # For every item in agency_name, shift one width to the right. Make the labels as vertical to avoid overlapping.
ax.legend(loc='upper left', ncols=2)
ax.set_ylim(0, 40000)

plt.show()
```

```python
# Hat graph
## Scenario: Game Theory

import matplotlib.pyplot as plt
import numpy as np

def hat_graph(ax, xlabels, values, group_labels): #ax means axes
    
    def label_bars(heights, rects): #rects means rectangles
    
        for height, rect in zip(heights, rects):
            #Attach a text label on top of each bar.
            ax.annotate(f'{height}',
                       xy=(rect.get_x() + rect.get_width() / 2, height),
                       xytext=(0, 4),  # 4 points vertical offset.
                       textcoords='offset points',
                       ha='center', va='bottom')
    
    values = np.asarray(values)
    x = np.arange(values.shape[1])
    ax.set_xticks(x, labels=xlabels)
    spacing = 0.3 # spacing between hat groups
    width = (1 - spacing) / values.shape[0]
    heights0 = values[0]
    for i, (heights, group_label) in enumerate(zip(values, group_labels)):
        style = {'fill': False} if i == 0 else {'edgecolor': 'black'}
        rects = ax.bar(x - spacing/2 + i * width, heights - heights0,
                       width, bottom=heights0, label=group_label, **style)
        label_bars(heights, rects)
        
xlabels = ['B1', 'B2', 'B3', 'B4', 'B5']
A1 = np.array([-1, -2, 1, 2, 0])
A2 = np.array([-2, 1, 0, -3, -1])
A3 = np.array([-3, -2, 0, 0, 0])

fig, ax = plt.subplots()
hat_graph(ax, xlabels, [A1, A2, A3], ['Move A1', 'Move A2', 'Move A3'])

ax.set_xlabel('Moves of Player B')
ax.set_ylabel('Payoff')
ax.set_ylim(-5, 5)
ax.set_title('Payoff to Each Choice of Move of Player A and Each Choice of Move of Player B')
ax.legend()

fig.tight_layout()
plt.show()
```

```python
# Table Demo
## Scenario: Corporate Bond Yield

import matplotlib.pyplot as plt
import numpy as np

data = [[4.82, 4.90, 4.93, 5.20, 5.13],
        [0.08, 0.08, 0.09, 0.07, 0.07],
        [0.10, 0.12, 0.12, 0.08, 0.10],
        [0.14, 0.14, 0.15, 0.09, 0.12],
        [0.17, 0.16, 0.17, 0.11, 0.13]]

columns = ['%s 2024' % x for x in ('Jan', 'Feb', 'Mar', 'Apr', 'May')]
rows = ['Maturity of %.1f' % y for y in (2.5, 2.0, 1.5, 1.0, 0.5)]

values = np.arange(4.8, 5.6, 0.1)
value_increment = 1

colors = plt.cm.BuPu(np.linspace(0.1, 0.6, len(rows))) #defines the range of colours
n_rows = len(data)

index = np.arange(len(columns)) + 0.3 #index means the x-axis positions of bars, which is 0.3, 1.3, 2.3, etc.
bar_width = 0.4

# Initialise the vertical-offset for the stacked bar chart.
y_offset = np.zeros(len(columns)) #true for all bars

# Plot bars and create text labels for the table
cell_text = []
for row in range(n_rows):
    plt.bar(index, data[row], bar_width, bottom=y_offset, color=colors[row])
    y_offset = y_offset + data[row]
    cell_text.append(['%1.2f' % x for x in y_offset]) #cell texts as the cumulative heights

# Reverse colours and text labels to display the last value at the top, as it is at the bottom by default.
colors = colors[::-1]
cell_text.reverse()

# Add a table at the bottom of the Axes
the_table = plt.table(cellText=cell_text,
                      rowLabels=rows,
                      rowColours=colors,
                      colLabels=columns,
                      loc='bottom')

# Adjust layout to make room for the table:
plt.subplots_adjust(left=0.2, bottom=0.2)

plt.ylabel('Monthly Average Spot Rates, Percent')
plt.yticks(values * value_increment, ['%.1f' % val for val in values]) #y labels
plt.xticks([])
plt.ylim(bottom=min(values))
plt.title('The Treasury High Quality Market (HQM) Corporate Bond Yields')

plt.show()
```

```python
# Bar of Pie
## Scenario: Investment in Hedge Funds

import matplotlib.pyplot as plt
import numpy as np

from matplotlib.patches import ConnectionPatch

# make figure and assign axis objects
fig, (ax3, ax1, ax2) = plt.subplots(1, 3, figsize=(12, 5))
fig.subplots_adjust(wspace=0)

# pie chart parameters
overall_ratios = [.50, .25, .02, .03, .05, .15]
labels = ['Funds', 'Arbitrage', 'Special situations', 'Distressed securities', 'Market neutral', 'Global macro']
explode = [0.1, 0.1, 0, 0, 0, 0] #the extent of the sector being "seperated" from the bar
# rotate so that first wedge(sector) is split fifty-fifty by the x-axis
angle = -180 * overall_ratios[0]
wedges, *_ = ax1.pie(overall_ratios, autopct = '%d%%', startangle = angle,
                     labels = labels, explode = explode)

# bar chart 1 parameters
fund_ratios = [.22, .14, .64]
fund_labels = ['Funds of funds', 'Sector funds', 'Global hedge funds']
bottom = 1
width = .2

# Adding from the top matches the legend.
for j, (height, label) in enumerate(reversed([*zip(fund_ratios, fund_labels)])):
    bottom -= height
    bc = ax2.bar(0, height, width, bottom=bottom, color='C0', label=label,
                 alpha=0.1 + 0.25 * j) #alpha means the transparency and j is an integer starting from 0
    ax2.bar_label(bc, labels=[f"{height:.0%}"], label_type='center')
    
ax2.set_title('Funds Breakdown')
ax2.legend(loc='center left', bbox_to_anchor=(0.8, 0.5))
ax2.axis('off')
ax2.set_xlim(- 2.5 * width, 2.5 * width)

# Use ConnectionPatch to draw lines between the pie chart and the bar chart 1
theta1, theta2 = wedges[0].theta1, wedges[0].theta2
center, r = wedges[0].center, wedges[0].r
bar_height = sum(fund_ratios)

# Draw top connecting line
x = r * np.cos(np.pi / 180 * theta2) + center[0]
y = r * np.sin(np.pi / 180 * theta2) + center[1]
con = ConnectionPatch(xyA=(-width / 2, bar_height), coordsA=ax2.transData, #-width means connect the pie to the left edge of the bar
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color([0, 0, 0]) #black
con.set_linewidth(2)
ax2.add_artist(con)

# Draw bottom connecting line
x = r * np.cos(np.pi / 180 * theta1) + center[0]
y = r * np.sin(np.pi / 180 * theta1) + center[1]
con = ConnectionPatch(xyA=(-width / 2, 0), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color([0, 0, 0])
con.set_linewidth(2)
ax2.add_artist(con)

# bar chart 2 parameters
arb_ratios = [.24, .36, .16, .24]
arb_labels = ['Index arbitrage', 'Fixed income arbitrage', 'Convertible arbitrage', 'Risk arbitrage']
bottom = 1
width = .2

for j, (height, label) in enumerate(reversed(list(zip(arb_ratios, arb_labels)))):
    bottom -= height
    bc = ax3.bar(0, height, width, bottom=bottom, color = 'C1', label=label, alpha=0.4 + 0.2 * j)
    ax3.bar_label(bc, labels=[f"{height:.0%}"], label_type='center')

ax3.set_title('Arbitrage Breakdown')
ax3.legend(loc='center left', bbox_to_anchor=(-0.5, 0.5))
ax3.axis('off')
ax3.set_xlim(-2.5 * width, 2.5 * width)

# Use ConnectionPatch to draw lines between the pie chart and the bar chart 2
theta3, theta4 = wedges[1].theta1, wedges[1].theta2
center, r = wedges[1].center, wedges[1].r
bar_height = sum(arb_ratios)

# top line
x = r * np.cos(np.pi / 180 * theta3) + center[0]
y = r * np.sin(np.pi / 180 * theta3) + center[1]

con = ConnectionPatch(
    xyA=(width / 2, bar_height), coordsA=ax3.transData, #+width means connect the pie to the right edge of the bar
    xyB=(x, y), coordsB=ax1.transData
)
con.set_color([0, 0, 0])
con.set_linewidth(2)
ax3.add_artist(con)

# bottom line
x = r * np.cos(np.pi / 180 * theta4) + center[0]
y = r * np.sin(np.pi / 180 * theta4) + center[1]

con = ConnectionPatch(
    xyA=(width / 2, 0), coordsA=ax3.transData,
    xyB=(x, y), coordsB=ax1.transData
)
con.set_color([0, 0, 0])
con.set_linewidth(2)
ax3.add_artist(con)

ax1.set_title('Investments Strategies Breakdown')

plt.show()
```

```python
# Nested pie charts (two methods)
##Method 1: using 'ax.pie'.

import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

size = 0.4
vals = np.array([[32., 48., 26.], [25., 71., 12.], [31., 20., 19.], [42., 15., 19.]])
outer_colors = ["#0033FF", "#AA1122", "#00FF55", "#885500"]
inner_colors = ["#0022EE", "#0011DD", "#0000CC", "#990011", "#880000", "#770000", "#00EE44", "#00DD33", "#00CC22",
                "#774400", "#663300", "#552200"]
outer_labels = ['Market A', 'Market B', 'Market C', 'Market D']
inner_labels = ['Product A', 'Product B', 'Product C', 'Product A', 'Product B', 'Product C', 'Product A', 
                'Product B', 'Product C', 'Product A', 'Product B', 'Product C']

ax.pie(vals.sum(axis = 1), radius = 1.2, colors = outer_colors, labels = outer_labels,
       wedgeprops = dict(width = size, edgecolor = 'w'))
ax.pie(vals.flatten(), radius = 1.2-size, colors = inner_colors, labels = inner_labels,
       wedgeprops = dict(width = size, edgecolor='w'))

ax.set(aspect="equal", title='Proportions of products in different markets')

plt.show()
```

```python
##Method 2: using 'ax.bar' and polar coordinates.

import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(subplot_kw=dict(projection="polar"))

size = 0.4
vals = np.array([[32., 48., 26.], [25., 71., 12.], [31., 20., 19.], [42., 15., 19.]])
# Normalise vals to 2 pi
valsnorm = vals/np.sum(vals)*2*np.pi
# Obtain the ordinates of the bar edges
valsleft = np.cumsum(np.append(0, valsnorm.flatten()[:-1])).reshape(vals.shape)

outer_colors = ["#0033FF", "#AA1122", "#00FF55", "#885500"]
inner_colors = ["#0022EE", "#0011DD", "#0000CC", "#990011", "#880000", "#770000", "#00EE44", "#00DD33", "#00CC22",
                "#774400", "#663300", "#552200"]
outer_labels = ['Market A', 'Market B', 'Market C', 'Market D']
inner_labels = ['Product A', 'Product B', 'Product C', 'Product A', 'Product B', 'Product C', 'Product A', 
                'Product B', 'Product C', 'Product A', 'Product B', 'Product C']

ax.bar(x = valsleft[:, 0], #obtain all rows and the first column of valsleft, which are 32, 25, etc.
       width = valsnorm.sum(axis = 1), bottom = 1.2-size, height = size,
       color = outer_colors, label = outer_labels, edgecolor = 'w', linewidth = 1, align = "edge")

ax.bar(x=valsleft.flatten(),
       width = valsnorm.flatten(), bottom = 1.2-2*size, height = size,
       color = inner_colors, label = inner_labels, edgecolor='w', linewidth = 1, align = "edge")

ax.set(title='Proportions of products in different markets')
ax.set_axis_off()
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5)) #Labels are not shown directly in ax.bar. Put the centre-left side of the legend box at the coordinate (1, 0.5).

plt.show()
```

```python
# Hatch demo

import matplotlib.pyplot as plt
import numpy as np

from matplotlib.patches import Ellipse, Polygon

x = np.arange(1, 5)
y1 = np.arange(1, 5)
y2 = np.ones(y1.shape) * 4

fig = plt.figure()
axs = fig.subplot_mosaic([['bar1', 'patches'], ['bar2', 'patches']])

axs['bar1'].bar(x, y1, edgecolor = 'black', hatch = '/')
axs['bar1'].bar(x, y2, bottom = y1, edgecolor = 'black', hatch = '///')

axs['bar2'].bar(x, y1, edgecolor = 'black', hatch = ['x', '|', '//', '+'])
axs['bar2'].bar(x, y2, bottom = y1, edgecolor = 'black', hatch = ['o', '*', 'O', '.'])

x = np.arange(0, 40, 0.2)
axs['patches'].fill_between(x, np.sin(x) * 4 + 30, y2 = 0,
                            hatch='///', zorder= 2 , fc = 'c')
axs['patches'].add_patch(Ellipse((4, 50), 10, 10, fill = True,
                                 hatch = '*', facecolor = 'y'))
axs['patches'].add_patch(Polygon([(10, 20), (30, 50), (50, 10)],
                                 hatch='\/...', facecolor='g'))
axs['patches'].set_xlim([0, 40])
axs['patches'].set_ylim([10, 60])
axs['patches'].set_aspect(1) #aspect ratio = 1

plt.show()
```
